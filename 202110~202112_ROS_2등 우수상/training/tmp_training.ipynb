```python
import os
import datetime
# DATAS_UP_PATH='../datas/output_up_camera'
# DATAS_DOWN_PATH='../datas/output_down_camera'
DATAS_UP_PATH='./DDARAGA/output_up_camera/'
DATAS_DOWN_PATH='./DDARAGA/output_down_camera/'
```

```python

d_time = datetime.datetime.now().strftime("%Y%m%d%H%M")
```

```python
file_list = os.listdir(DATAS_DOWN_PATH)
for file_name in file_list:
    if file_name[-4:] == '.jpg':
        os.rename(DATAS_DOWN_PATH+file_name,DATAS_DOWN_PATH+file_name[:-4]+'_'+d_time+'.jpg')
        print(file_name)
file_list = os.listdir(DATAS_UP_PATH)
for file_name in file_list:
    if file_name[-4:] == '.jpg':
        os.rename(DATAS_UP_PATH+file_name,DATAS_UP_PATH+file_name[:-4]+'_'+d_time+'.jpg')
        print(file_name)
```

```python
up_list = os.listdir(DATAS_UP_PATH)
down_list = os.listdir(DATAS_DOWN_PATH)
```

```python
for up in up_list:
    if up not in down_list:
        os.remove(DATAS_UP_PATH+up)
        print(up)
```

```python
for down in down_list:
    if down not in up_list:
        os.remove(DATAS_DOWN_PATH+down)
        print(down)
```
